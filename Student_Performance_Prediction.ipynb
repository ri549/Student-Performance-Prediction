```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
data['Extracurricular Activities']=le.fit_transform(data['Extracurricular Activities'])
```

```python
X = data.drop('Performance Index', axis=1)
```

```python
X
```

```python
Y=data['Performance Index']
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=50)
```

```python
x_train.shape,y_train.shape,x_test.shape,y_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
x_train_scaler=scaler.fit_transform(x_train)
x_test_scaler=scaler.transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
model1=LinearRegression()
```

```python
model1_params = {
    'fit_intercept': [True, False]
}
```

```python
model1_grid = GridSearchCV(model1,model1_params,cv=5,scoring='r2')
```

```python
model1_grid.fit(x_train_scaler,y_train)
```

```python
best_model = model1_grid.best_estimator_
```

```python
y_pred1=model1_grid.predict(x_test_scaler)
```

```python
from sklearn.metrics import mean_squared_error, r2_score  , mean_absolute_error, root_mean_squared_error
from sklearn.metrics import confusion_matrix
```

```python
mse=mean_squared_error(y_test,y_pred1)
r2=r2_score(y_test,y_pred1)
mae=mean_absolute_error(y_test,y_pred1)
rmse=np.sqrt(mse)
```

```python
print(f"mean_squared_error: {mse}")
print(f"r2_score: {r2}")
print(f"mean_absolute_error: {mae}")
print(f"root_mean_squared_error: {rmse}")
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
model2=DecisionTreeRegressor()
```

```python
model2_params = {
    'max_depth': [None, 5, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}
```

```python
model2_grid=GridSearchCV(model2,model2_params,cv=5,scoring='r2')
```

```python
model2_grid.fit(x_train_scaler,y_train)
```

```python
y_pred2=model2_grid.predict(x_test_scaler)
```

```python
mse=mean_squared_error(y_test,y_pred2)
r2=r2_score(y_test,y_pred2)
mae=mean_absolute_error(y_test,y_pred2)
rmse=np.sqrt(mse)
```

```python
print(f"mean_squared_error: {mse}")
print(f"r2_score: {r2}")
print(f"mean_absolute_error: {mae}")
print(f"root_mean_squared_error: {rmse}")
```

```python
# Create a DataFrame to compare actual vs. predicted for each student
individual_performance = pd.DataFrame({
    'Student_ID': y_test.index,
    'Actual_Performance': y_test,
    'Predicted_Performance': y_pred1
})

# Add a column for the prediction error (difference)
individual_performance['Prediction_Error'] = abs(individual_performance['Actual_Performance'] - individual_performance['Predicted_Performance'])

# Display the results
print(individual_performance)
```
